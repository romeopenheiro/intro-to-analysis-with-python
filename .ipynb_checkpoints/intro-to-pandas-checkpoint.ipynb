```python
city_pop_file = "https://milliams.com/courses/data_analysis_python/city_pop.csv"
pd.read_csv(city_pop_file)
```

```python
pd.read_csv(
    city_pop_file,
    skiprows=5,
)
```

```python
pd.read_csv(
    city_pop_file,
    skiprows=5,
    sep=";",
)
```

a value of -1 signifies a missing reading so we should mark those too.

```python
pd.read_csv(
    city_pop_file,
    skiprows=5,
    sep=";",
    na_values="-1",
)
```

use the year column as the index for the DataFrame

```python
census = pd.read_csv(
    city_pop_file,
    skiprows=5,
    sep=";",
    na_values="-1",
    index_col="year",
)
census
```
