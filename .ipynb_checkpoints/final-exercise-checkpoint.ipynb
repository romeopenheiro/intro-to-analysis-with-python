```python
all_males = titanic[(titanic["gender"] == "male")]
all_females = titanic[(titanic["gender"] == "female")]

survived_males = titanic[(titanic["gender"] == "male") & (titanic["survived"] == "yes")]
survived_females = titanic[(titanic["gender"] == "female") & (titanic["survived"] == "yes")]
```

```python
percent_male = (survived_males["survived"].count() / all_males["gender"].count()) * 100
percent_male
```

```python
percent_female = (survived_females["survived"].count() / all_females["gender"].count()) * 100
percent_female
```

```python
def survived_ratio(dataframe):
    yes = dataframe[dataframe["survived"] == "yes"]
    return (len(yes)/ len(dataframe))*100
```

```python
ratio_1st = survived_ratio(titanic[titanic["class"] == "1st"])
ratio_2nd = survived_ratio(titanic[titanic["class"] == "2nd"])
ratio_3rd = survived_ratio(titanic[titanic["class"] == "3rd"])

print(ratio_1st, ratio_2nd, ratio_3rd)
```

```python
male_ratio = survived_ratio(titanic[titanic["gender"] == "male"])
female_ratio = survived_ratio(titanic[titanic["gender"] == "female"])

print(male_ratio, female_ratio)
```
