```python
import seaborn as sns
import matplotlib.style as style 
style.use('seaborn-colorblind')
```

```python
sns.set_theme()
```

```python
tips = pd.read_csv("https://milliams.com/courses/data_analysis_python/tips.csv")
```

```python
sns.relplot(data = tips, x="total_bill", y="tip")
```

```python
g = sns.relplot(data = tips, x="total_bill", y="tip")
g.set_axis_labels("Total bill (£)", "Tip (£)")
```

```python
sns.relplot(data=tips, x="total_bill", y="tip", hue="day", size="size", style="time")
```

```python
sns.relplot(data=tips, x="total_bill", y="tip", hue="day", style="day")
```

```python
tips["percent_tip"] = (tips["tip"] / tips["total_bill"])*100
tips["bill per person"] = tips["total_bill"] / tips["size"]
```

```python
sns.relplot(data=tips, x="bill per person", y="percent_tip", hue="day", style="day")
```

```python
sns.displot(tips, x = "total_bill")
```

```python
sns.displot(tips, x = "total_bill", col = "time")
```

```python
sns.displot(tips, x = "total_bill", col = "time", stat = "density")
```

```python
sns.displot(tips, x = "total_bill", col = "time", stat = "density", common_norm = False)
```

```python
sns.displot(data=tips, x="total_bill", hue="time", stat="density", common_norm=False)
```

```python
# applying a "kernel density estimation" which allows you to turn it into a line graph:
sns.displot(data=tips, x="total_bill", hue="time", kind="kde", common_norm=False)
```
